```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcdefaults()
import matplotlib as mpl
mpl.rc("figure", facecolor="white")
plt.rc('axes', axisbelow=True)
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from matplotlib import cm as cm
import warnings
from scipy.stats import pearsonr
warnings.filterwarnings('ignore')
```

# Exploratory Data Analysis

```python
# we will consider the following 7 genes:
immune_genes = ['CD247', 'CD2', 'CD3E', 'GZMH', 'NKG7', 'PRF1', 'GZMK']

data = pd.read_csv('../data.tsv', sep='\t')
data.head(3)
```

```python
# log of FPKM
data[immune_genes].plot(kind='box', figsize=(12,8), logy=True)
#plt.title('Immune Signature- Individual Gene Expression', fontsize=26)
plt.ylabel('log(FPKM)', fontsize=24)
plt.yticks(fontsize=24)
plt.xticks(fontsize=24)
plt.grid()
plt.show()
```

```python
# Distribution of immune sig scores
data['immune_signature'].plot(kind='hist', edgecolor='black', bins=np.arange(0,1.01,0.035), figsize=(12,8))
plt.xlabel('Immune Signature', fontsize=24)
plt.ylabel('Count', fontsize=24)
#plt.title('Immune Signature: TCGA-SKCM', fontsize=26)
plt.xticks(np.arange(0,1.001, 0.1), fontsize=22)
plt.yticks(fontsize=22)
plt.xlim([0, 1])
plt.grid()
plt.show()
```

```python
# Distribution of non-silent mutations
data['non_silent_muation'].plot(kind='hist',
                                edgecolor='black',
                                figsize=(12,8),
                                bins=np.arange(0,data['non_silent_muation'].max(),5))
plt.xlabel('Non-Silent Mutation Rate (mutations/Mb exome)', fontsize=24)
plt.ylabel('Count', fontsize=24)
#plt.title('Non-Silent Mutation Rate: TCGA-SKCM', fontsize=26)
plt.xlim([0, data['non_silent_muation'].max()+1])
plt.grid()
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.show()
```

```python
# ZOOM IT IN
# Distribution of non-silent mutations
data['non_silent_muation'].plot(kind='hist',
                                edgecolor='black',
                                figsize=(12,8),
                                bins=np.arange(0,data['non_silent_muation'].max(),1))
plt.xlabel('Non-Silent Mutation Rate (mutations/Mb exome)', fontsize=24)
plt.ylabel('Count', fontsize=24)
#plt.title('Non-Silent Mutation Rate: TCGA-SKCM', fontsize=26)
plt.xlim([0, data['non_silent_muation'].max()+1])
plt.grid()
plt.xlim([0,50])
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.show()
```

```python
# mutation rate vs immune signature
data.plot(x='immune_signature',
          y='non_silent_muation',
          kind='scatter',
          s=60,
          alpha=0.4,
          edgecolor='black',
          c='salmon',
          figsize=(12,8))
plt.ylabel('Non-Silent Mutation Rate\n (mutations/Mb exome)', fontsize=24)
plt.xlabel('Immune Signature', fontsize=24)
#plt.title('Non-Silent Mutation Rate vs Immune Signature', fontsize=26)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

# pearsons correlation coefficient
correlation = pearsonr(x=data['immune_signature'], y=data['non_silent_muation'])
plt.text(0.43, 305, 'Pearson\'s r = {0}'.format(round(correlation[0], 2)), fontsize=24, color='darkgreen')
plt.text(0.43, 275, 'p = {0}'.format(round(correlation[1], 2)), fontsize=24, color='darkgreen')



# add a best fit line
plt.plot(np.unique(data['immune_signature']),
         np.poly1d(np.polyfit(data['immune_signature'], data['non_silent_muation'], 1))(np.unique(data['immune_signature'])))

plt.xlim([0, 1])
plt.ylim([0, data['non_silent_muation'].max()+5])
plt.grid()
plt.show()
```
